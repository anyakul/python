```python
f, ff = sympy.symbols('f ff')
f = ((-12 * (x ** 4)) * (sin(cos(x))) - (18 * (x ** 3)) + (5 * (x ** 2)) + 10 * x - 30)
```

```python
ff = sympy.diff(f)
ff
plot(f, ff)
```

Функция убывает на интервалах х ∈ (-10 ; -7] и х ∈ [4 ; 7]

f > 0 на промежутках х ∈ (-∞ ; -7,5), х ∈ [-5 ; -2,5], х ∈ [2,5 ; 4], х ∈ [8 ; +∞)
