```python
from math import sqrt

def solve(a, b, c):
    d = b**2 - 4*a*c
    if d >= 0:
        x1 = (-b + sqrt(d)) / (2 * a)
        x2 = (-b - sqrt(d)) / (2 * a)
        return (x1, x2)
    else:
        return 'Вещественных корней нет'
solve(0.5, 0.125, 0)
```

```python
solve(1, 2, 3)
```

```python
import cmath

def csolve(a, b, c):
    d = b**2 - 4*a*c
    x1 = (-b + cmath.sqrt(d)) / (2 * a)
    x2 = (-b - cmath.sqrt(d)) / (2 * a)
    return (x1, x2)
csolve(1, 2, 3)
```

```python
import matplotlib.pyplot as plt
import random

vx = [1.0, 2.0, 3.0]
vy = [1.0, 2.0, 1.0]

sx = vx[0]
sy = vy[0]

px = []
py = []

for i in range(1, 1000):
    r = random.randint(0, 2)
    sx = (sx + vx[r]) / 2
    sy = (sy + vy[r]) / 2
    px.append(sx)
    py.append(sy)

plt.plot(px, py, 'o')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

count = 100
x = list(range(1, 7))
y = [0 for i in x]

fig, ax = plt.subplots()

for _ in range(count):
    y[random.randint(0, 5)] += 1

ax.bar(x, y)

y.sort()

stat = [f'{i+1}: {round(y[i] / count * 100)}%' for i in range(0, 6)]
print(stat)
plt.show()
```
