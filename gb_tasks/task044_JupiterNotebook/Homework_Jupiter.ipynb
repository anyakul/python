```python
x = np.arange(-20,20.1,0.01)
fx=0*x
npf = ((-12)**4)*np.sin(np.cos(x))-18*x**3+5*x**2+10*x-30 #Функция
npff = -54*x**2+10*x-20736*np.sin(x)*np.cos(np.cos(x))+10 #Производная
plt.plot(x,npf)
plt.plot(x,npff)
plt.plot(x,fx)
plt.show()
```

```python
vershiny = []
i = -20
di=0.01
lastdf = -54*i**2+10*i-20736*np.sin(i)*np.cos(np.cos(i))+10

while i < 20:
    i+=di    
    nextdf = -54*i**2+10*i-20736*np.sin(i)*np.cos(np.cos(i))+10

    if nextdf*lastdf<0:
        vershiny.append(i)
      
    lastdf = nextdf

for el in vershiny:
    print(f"({round(el,1)}, {round(((-12)**4)*np.sin(np.cos(el))-18*el**3+5*el**2+10*el-30, 1)})")
```
